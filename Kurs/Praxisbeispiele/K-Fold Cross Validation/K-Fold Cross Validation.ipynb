```python
import pandas as pd

# nur die ersten 6 Zeilen laden

df = pd.read_csv("hotels.csv")

df
```

```python
X = df[["Gewinn", "Quadratmeter"]].values
Y = df[["Preis in Mio"]].values
```

```python
X
```

```python
# Gibt den Eintrag an Stelle 0 und 1 zurück
X[[0,1]]
```

```python
from sklearn.model_selection import KFold
from sklearn.linear_model import LinearRegression

# n_splits gibt die Anzahl der Aufteilungen wieder
# shuffle gibt an, ob die Daten durchmischt werden sollen
kf = KFold(n_splits = 10, shuffle = True)

for train_index, test_index in kf.split(X):
    print("train: " + str(train_index))
    print("test: " + str(test_index))
    print("----------------")
    X_test = X[test_index]
    X_train = X[train_index]
    
    Y_test = Y[test_index]
    Y_train = Y[train_index]
    
    model = LinearRegression()
    model.fit(X_train,Y_train)
    
    print(model.score(X_test, Y_test))
```
