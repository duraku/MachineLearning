```python
import pandas as pd

df = pd.read_csv("classification.csv")

df.head()
```

```python
from sklearn.model_selection import train_test_split

X = df[["age", "interest"]].values
Y = df["success"].values

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, random_state = 0, test_size = 0.25)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X_train)

X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.svm import SVC
import numpy as np 

# gamma Parameter hoch bedeutet hoher Berg der sehr schnell abf√§llt
# passt sich sehr stark den Date an
model = SVC(kernel = "rbf", gamma = 0.5, C = 1)
model.fit(X_train, Y_train)

model.score(X_test, Y_test)
```

```python
from helper import plot_classifier

plot_classifier(model, X_train, Y_train, proba = False, xlabel = "Alter", ylabel = "Interesse")
```
