```python
import pandas as pd

df = pd.read_csv("classification.csv")

df.head()
```

```python
from sklearn.model_selection import train_test_split

X = df[["age", "interest"]].values
Y = df["success"].values

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, random_state = 0, test_size = 0.25)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X_train)

X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.svm import SVC
import numpy as np 

# (u' * v + coef0)^degree
# coef0 auf 1 setzten  und degree erhöhen (Max 2-3) für mehr Flexibilität
# kann aber zur Überanpassung führen
model = SVC(kernel = "poly", degree = 2, coef0 = 1)
model.fit(X_train, Y_train)

model.score(X_test, Y_test)
```

```python
from helper import plot_classifier

plot_classifier(model, X_train, Y_train, proba = False, xlabel = "Alter", ylabel = "Interesse")
```
